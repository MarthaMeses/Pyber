```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# File to Load
fileCity_csv = "../Resources/city_data.csv"
fileRide_csv = "../Resources/ride_data.csv"
```

```python
# Read the City and Ride Data
dataCity_df = pd.read_csv(fileCity_csv)
dataRide_df = pd.read_csv(fileRide_csv)
```

```python
dataCity_df.info()
```

```python
dataCity_df
```

```python
dataRide_df.info()
```

```python
# Combine the data into a single dataset
# using an inner join
pyber_df = pd.DataFrame()
pyber_df = pd.merge(dataRide_df, dataCity_df, on="city")
pyber_df
```

```python
pyber_df.info()
```

```python
pyber_df.sort_values(by=['city'], inplace=True)
pyber_df.head()
```

```python
# Display the data table for preview/ NOTEBOOK EXAMPLE 
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
# Create data
```

```python
cityTypes = pyber_df.groupby('type')['city'].count()
cityTypes
```

```python
totalNumbersRides = pyber_df.groupby(['type','city'])['ride_id'].count()
totalNumbersRides
```

```python
averageFare =  pyber_df.groupby(['type','city'])['fare'].mean()
averageFare
```

```python
driverCity = pyber_df.groupby(['type','city'])['driver_count'].count()
driverCity
```

```python
# Build the scatter plots for each city types
plt.scatter(totalNumbersRides.Urban, averageFare.Urban, alpha=0.4, s=driverCity*10, c='lightcoral', label='Urban', edgecolors="red", linewidth=2)
plt.scatter(totalNumbersRides.Suburban, averageFare.Suburban, alpha=0.5, s=driverCity*10, c='lightblue', label='Suburban', edgecolors="blue", linewidth=2)
plt.scatter(totalNumbersRides.Rural, averageFare.Rural, alpha=0.6, s=driverCity*10, c='gold', label='Rural', edgecolors="orange", linewidth=2)

# Incorporate the other graph properties
plt.grid()
# Create a legend
plt.title('Pyber Ride Sharing Data')
plt.xlabel('Total Numbers of Rides (per City)')
plt.ylabel('Average Fare ($)')
# Incorporate a text label regarding circle size
plt.legend(loc=1)
plt.show()
```

```python
# Save Figure
plt.savefig('pyberRideData.png')
```

## Total Fares by City Type

```python
# Calculate Type Percents
percentFare =  pyber_df.groupby(['type'])['fare'].sum()
percentFare
```

```python
# Build Pie Chart
city_type = ["Rural", "Suburban", "Urban"]
x_axis = np.arange(0, len(city_type))
colors = ["gold", "lightskyblue", "lightcoral",]
explode = (0,0,0.05)
plt.pie(percentFare, explode=explode, labels=city_type, colors=colors, autopct="%1.1f%%", shadow=True, startangle=140)
plt.axis("equal")
plt.title('% of Total Fare by City Type')
```

```python
# Show Figure
plt.show()
```

```python
# Save Figure
plt.savefig('totalFareByCityType.png')
```
